# 2 . Ordinal Encoder

```python
df=pd.read_csv("C:\\Users\\Lenovo\\Documents\\Summer internship\\Machine Learning\\covid_toy.csv")
```

```python
df=df.drop(columns=['age','fever'])
```

```python
df.head()
```

```python
df['city'].value_counts()
```

```python
df['cough'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Male','Female'],['Mild','Strong'],['Kolkata','Bangalore','Delhi','Mumbai'],['Yes','No']])
```

```python
oe
```

```python
oe.fit(df)
```

```python
df_new=oe.transform(df)
```

```python
oe.categories_
```

```python
df=pd.DataFrame(df_new,columns=df.columns)
```
